# List of core section and facies
I create a list having core section and facies from the XRF database  for Dirk to pick cores for detailed classification.

```python
import pandas as pd
import os
os.getcwd()
```

```python
info_df = pd.read_csv('data/info.cleaned.all.csv', index_col=0)
info_df.columns
```

```python
labels = pd.read_csv('data/XRF_y_label.csv', index_col=0)
labels
```

```python
compile_df = pd.concat([info_df['core_section'], labels], join='inner', axis=1)
compile_df
```

```python
facies = ['shoreface', 'channel', 'beach-foreshore', 'sand flat', 'mud flat',
       'lagoon', 'peat', 'soil', 'eolian/fluvial (W)',
       'shallow marine', 'moraine', 'eolian/fluvial']

Facies = []
for fa in facies:
    Facies.append(fa.capitalize())

Facies
```

## Method 1: crate a dictionary

```python
section_dict = {'facies': 'core sections'}
for facies in compile_df.facies_merge_2.unique():
    #section_list.append(compile_df.loc[compile_df.facies_merge_2 == facies, 'core_section'].unique())
    section_dict[Facies[facies]] = compile_df.loc[compile_df.facies_merge_2 == facies, 'core_section'].unique()

section_dict
```

```python
import csv

with open('results/facies-coresection_list.csv', 'w') as f:
    w = csv.writer(f)
    for key, val in section_dict.items():
        w.writerow([key, val])
```

Mmm...the output is not suitable for excel...

## Method 2: list and dataframe

```python
section_list = []
facies_list = []

for facies in compile_df.facies_merge_2.unique():
    section_list.append(compile_df.loc[compile_df.facies_merge_2 == facies, 'core_section'].unique())
    facies_list.append(Facies[facies])
```

```python
pd.DataFrame(section_list, index=facies_list).T.to_csv('results/facies-coresection_list.csv')
```

# Add a column to legends
The added column is the capitalized facies name followed "new_label" in the facies_legends.csv.

```python
legends = pd.read_csv('data/facies_legends.csv', index_col=0)
legends
```

```python
new_facies = []
for facies in legends.new_label:
    if facies > 11:
        new_facies.append(None)
    else:
        new_facies.append(Facies[facies])
    
legends['new_facies'] = new_facies
legends
```

```python
legends.to_csv('data/facies_legends.csv')
```
