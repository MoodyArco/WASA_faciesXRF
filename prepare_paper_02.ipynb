# Update the core list

```python
from wasafacies import PrepareData

prepare = PrepareData()
facies, id_list = prepare.create_recla()

info_df = pd.read_csv('data/info.cleaned.all.csv', index_col=0, 
                      usecols=['composite_id', 'core_section', 'long_utm', 'lat_utm'])
```

```python
info_df
```

```python
out = info_df.loc[id_list, ['core_section', 'long_utm', 'lat_utm']].groupby('core_section').mean()
out
```

```python
import utm
lat, lon = utm.to_latlon(out.long_utm.values, out.lat_utm.values, 32, 'U')

out['lon'] = lon
out['lat'] = lat

out
```
